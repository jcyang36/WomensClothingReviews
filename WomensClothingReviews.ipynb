```python


import nltk
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.tokenize import TweetTokenizer
from nltk.probability import FreqDist
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer

from nltk.corpus import names

from string import punctuation
```

```python
filepath="datasets/women_clothing_review.csv"
```

```python
import pandas as pd
```

```python
df=pd.read_csv(filepath,encoding="latin-1")

df.head()

df.dropna()
```

```python
sid=SentimentIntensityAnalyzer()
```

```python
eng_stopwords = stopwords.words('english')
eng_stopwords
```

```python

def cleanReview(review):
    review=str(review).lower()
    tknz_review=word_tokenize(review)
    
    #remove punctuation
    for token in tknz_review:
        if token in punctuation:
            tknz_review.remove(token)
            
    clean_tokens=[]
    
    #remove filler words
    for token in tknz_review:
        if token in eng_stopwords:
            if token in ['not','but',"don't","aren't","couldn't","doesn't","didn't","hadn't","hasn't","isn't","wasn't","weren't","won't","wouldn't"]:
                clean_tokens.append(token)
        if token not in eng_stopwords:
            clean_tokens.append(token)
    
    #put sentence back together with remaining clean words
    clean_review=' '.join(clean_tokens)
    return clean_review
    
def getReviewSentiment(clean_review):
    sid_rev=sid.polarity_scores(clean_review)
    
    r_comp=sid_rev['compound']
    return r_comp

def analyzeReview(review):
    return getReviewSentiment(cleanReview(review))
```

```python
df['review_sentiment']=df['Review Text'].apply(analyzeReview)
```
